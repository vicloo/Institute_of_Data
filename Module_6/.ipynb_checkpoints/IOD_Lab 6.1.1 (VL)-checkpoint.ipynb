### 3. Prepare the data for clustering

1. Remove the `sex`, `sport` column. We will see if the clusters from K-Means end up like the actual sex.
2. Put the features on the same scale.

```python
# ANSWER
X = ais.drop(['sex', 'sport'], axis = 1)
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

### 4. Clustering with K-Means

- Cluster the data to our our target groups.
    - Use sklearn.cluster.KMeans
- We know that there are 2 actual classes. Lets try using 2, 4. Run the clustering algorithm with each of those k values.

```python
# ANSWER
from sklearn.cluster import KMeans

k=2

k_mean = KMeans(n_clusters=k)
k_mean.fit(X_scaled)
```

### 5. Get the labels for the models

```python
# ANSWER - Labels for 2 clusters
k_mean.labels_
```

### 6. Elbow Method For Optimal k

For each k value, we will initialise k-means and use the inertia attribute to identify the sum of squared distances of samples to the nearest cluster centre. As k increases, the sum of squared distance tends to zero. 

```python
k_mean.inertia_
```

```python
# ANSWER
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(X_scaled)
    Sum_of_squared_distances.append(km.inertia_)
```

Below is a plot of sum of squared distances for k in the range specified above. If the plot looks like an arm, then the elbow on the arm is optimal k.

```python
# ANSWER
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum of squared distances')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
markers = ['o', 'v', 's', 'H']
colors = ['b', 'g', 'c', 'y']

centers = np.array(kmeans_model.cluster_centers_)

sns.lmplot(x='pcBfat', y='wt', data=ais_labels, hue='labels', markers=markers, palette=colors, fit_reg=False);
plt.scatter(centers[:, 7], centers[:, 10], marker='x', color='r')

plt.title('k means centroids')
plt.xlabel('pcBfat')
plt.ylabel('wt')
```
